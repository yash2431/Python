### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 15, 7, 12, 5]
plt.bar(categories, values, color='skyblue')
plt.title('Bar Chart with Values')
plt.xlabel('Categories')
plt.ylabel('Values')
for i, value in enumerate(values):
    plt.text(i, value + 0.5, str(value), ha='center', va='bottom')
plt.show()
```

### 09) WAP create a Scatter Plot with several colors in Matplotlib?

```python
import matplotlib.pyplot as plt
x= [8,9,6,3,5,7,4,1,2,6]
y= [5,8,9,6,3,2,4,8,8,9]
plt.scatter(x, y, c='r', alpha=0.5)
plt.colorbar(label='Color Scale')
plt.title('Scatter Plot with Several Colors')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### 10) WAP to create a Box Plot.

```python
import matplotlib.pyplot as plt
x= [8,9,6,3,5,7,4,1,2,6]
y= [5,8,9,6,3,2,4,8,8,9]
plt.boxplot(x)
plt.boxplot(y)
plt.title('Box Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
